```python
#pandas is a library and it contains the functions related to the database
import pandas as pd


#import pandas for creating dataframe to perform CRUD operation
```

```python

dataset=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Machine Learning-Regression/1.Traditional Way/1.Simple Linear Regression/Salary_Data.csv")#dataset is
#to read the csv from the drive
```

```python
dataset
#to dataset structure
```

```python
independent=dataset[["YearsExperience"]]#
independent

#to store independent value from dataset
```

```python
dependent=dataset[["Salary"]]
dependent

#to store dependent value from dataset
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)

#to split the train test split
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)

#to import linear regressor and train  that model
```

```python
weight=regressor.coef_
weight

#to see the weights
```

```python
bais=regressor.intercept_
bais

#to see the bias
```

```python
y_pred=regressor.predict(X_test)
#to make prdiction for test data
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)

#evaluating model using r2score
```
