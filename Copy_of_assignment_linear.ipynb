```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Machine Learning-Regression/Assignment/insurance_pre.csv")
```

```python
df
```

```python
df = pd.get_dummies(df,drop_first=True)

df
```

```python
x = df.drop(["charges"],axis=1)
y = df.charges
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

from sklearn.pipeline import Pipeline

from sklearn.linear_model import LinearRegression
```

```python
pipeline1 = Pipeline([
    ('scaler', StandardScaler()),   # Step 1: StandardScaler for preprocessing
    ('regressor', LinearRegression())  # Step 2: LinearRegression as the model
])
model1 = LinearRegression()

```

```python
pipeline1.fit(xtrain,ytrain)
model1.fit(xtrain,ytrain)
```

```python
ypred1 = pipeline1.predict(xtest)
ypred2 = model1.predict(xtest)
```

```python
from sklearn.metrics import r2_score
result1 = r2_score(ypred1,ytest)
result2 = r2_score(ypred2,ytest)
```

```python
print(result1,result2)
```

```python
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.fit_transform(xtest)
model2 = LinearRegression(n_jobs=10)
model2.fit(xtrain,ytrain)
model2.score(xtest,ytest)
```
