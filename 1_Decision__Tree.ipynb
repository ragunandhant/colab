```python
dataset=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Machine Learning-Regression/Assignment/insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset.drop(['charges'],axis=1)
```

```python
independent
```

```python
dependent=dataset[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='best')
regressor=regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='random',max_depth=20)
regressor=regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='best',max_depth=8)
regressor=regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='best',random_state=3,max_depth=5,max_leaf_nodes=3)
regressor=regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
regressor=DecisionTreeRegressor(criterion='poisson', splitter='best',max_depth=10,max_leaf_nodes=7)
regressor=regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score*100
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'criterion':['poisson'],
    'splitter':['best','random'],
    'max_depth': [5,6,7,8,9 ],  # Maximum depth of the tree
    'min_samples_split': [5,6,7,8,9],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [5,6,7,8,9]     # Minimum number of samples required to be at a leaf node
}
```

```python
grid_search = GridSearchCV(DecisionTreeRegressor(), param_grid, cv=5, scoring='r2', n_jobs=-1)

# Step 5: Perform the grid search on the training data
grid_search.fit(X_train, y_train)

# Step 6: Get the best hyperparameters and the best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```

```python
print(grid_search.cv_results_)
```

```python
grid_search.best_estimator_
```

```python
grid_search.score(X_test,y_test)
```
