```python
dataset=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Machine Learning-Regression/Assignment/insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset.drop(['charges'],axis=1)
```

```python
dependent=dataset[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
X_train
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='rbf', C=10000,epsilon=1000,degree=5)
regressor.fit(X_train,y_train)
regressor.fit_status_
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
regressor.support_
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python

from sklearn.model_selection import train_test_split, GridSearchCV




# Step 3: Define the hyperparameter grid for SVR
param_grid = {
    'kernel': ['rbf'],
    'C': [0.1, 1, 10,1000,10000],
    'epsilon': [0.01, 0.1, 0.2,1000,500,100]
}

# Step 4: Create the GridSearchCV object
grid_search = GridSearchCV(SVR(), param_grid, cv=5, scoring='r2', n_jobs=-1)

# Step 5: Perform the grid search on the training data
grid_search.fit(X_train, y_train)

grid_search.score(X_test,y_test)
```

```python
grid_search.best_estimator_
```
